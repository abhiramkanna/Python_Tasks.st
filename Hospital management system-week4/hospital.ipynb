```python
#creating the parent class as patients
class patients:
    def __init__(self,pat_id,pat_name,gender):
        self.pat_id=pat_id
        self.pat_name=pat_name
        self.gender=gender
        #encapsulation
        self.__disease ="Fever"
    def show_patients(self):
        print("patient_id:",self.pat_id)
        print("patient name:",self.pat_name)
        print("gender",self.gender)
        print("Disease :", self.__disease)
#creating child class as doctors(inheritance)
class Doctor(patients):
    def __init__(self, pat_id, pat_name, doc_name):
        super().__init__(pat_id, pat_name, doc_name)
        self.doc_name=doc_name
    def show_doctors(self):
        self.show_patients()
        print("doctor_name:",self.doc_name)

#creating the objects
p1 = patients(101, "Rahul","male")
p2 = patients(102, "Priya","female")
p3 = patients(103, "Rohan","male")
d1 = Doctor(102, "Arjun", "Dr. Sharma")
d2 = Doctor(104, "Sneha", "Dr. Reddy")
d3 = Doctor(101,"Rahul", "Dr. Shyam")
```

```python
#display the values patients
p1.show_patients()
p2.show_patients()
p3.show_patients()
#display the values doctors
d1.show_doctors()
d2.show_doctors()
d3.show_doctors()
```
